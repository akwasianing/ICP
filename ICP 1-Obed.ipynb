```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasets/hepatitis/master/data/hepatitis.csv')
df.head(10)
```

```python
#check for missing values
df.isnull().sum()
```

```python
# Identify numerical columns
numerical_columns = df.select_dtypes(include=['number']).columns
print(numerical_columns)
```

```python
#impute the mean for bilirubin (filling null values using mean)
bilirubin_mean = df['bilirubin'].mean()
bilirubin_mean = round(bilirubin_mean, 1)
print(bilirubin_mean)
df['bilirubin'] = df['bilirubin'].fillna(round(bilirubin_mean))
df['bilirubin'] = df['bilirubin'].round(1)
df['bilirubin'].isnull().sum()
```

```python
#impute the mean for alk_phosphate (filling null values using mean)
alk_phosphate_mean = df['alk_phosphate'].mean()
alk_phosphate_mean = round(alk_phosphate_mean, 1)
print(alk_phosphate_mean)
df['alk_phosphate'] = df['alk_phosphate'].fillna(round(alk_phosphate_mean))
df['alk_phosphate'] = df['alk_phosphate'].round(1)
df['alk_phosphate'].isnull().sum()
```

```python
#impute the mean for sgot (filling null values using mean)
sgot_mean = df['sgot'].mean()
sgot_mean = round(sgot_mean, 1)
print(sgot_mean)
df['sgot'] = df['sgot'].fillna(round(sgot_mean))
df['sgot'] = df['sgot'].round(1)
df['sgot'].isnull().sum()
```

```python
albumin_mean = df['albumin'].mean()
albumin_mean = round(albumin_mean, 1)
print(albumin_mean)
df['albumin'] = df['albumin'].fillna(round(albumin_mean))
df['albumin'] = df['albumin'].round(1)
df['albumin'].isnull().sum()
```

```python
#impute the mean for protime (filling null values using mean)
protime_mean = df['protime'].mean()
protime_mean = round(protime_mean, 1)
print(protime_mean)
df['protime'] = df['protime'].fillna(round(protime_mean))
df['protime'] = df['protime'].round(1)
df['protime'].isnull().sum()
```
